**Copy Data Pandas**

```python
data_copy = df_csv.copy()
data_copy.head()
```

```python
# Copy kolom tertentu
data_baru = df_csv[['PassengerId','Survived','Name','Sex']].copy()
data_baru.head()
```

**Add kolom Pandas**

```python
data_baru['total_baru'] = data_baru['PassengerId'] + 5
data_baru.head()
```

**Remove Kolom Pandas**

```python
data_baru = data_baru.drop(columns=['total_baru'])
data_baru.head()
```

**Conditional Changes pandas**

```python
data_baru.loc[data_baru['Survived'] == 1, 'Status'] = 'Selamat'
data_baru.loc[data_baru['Survived'] == 0, 'Status'] = 'Meninggal'
data_baru.head()
```

**Export/Download Data**

```python
# Dalam CSV
data_baru.to_csv('output/file_csv.csv', index=False)
```

```python
# Dalam Excel
data_baru.to_excel('output/file_excel.xlsx', index=False)
```
